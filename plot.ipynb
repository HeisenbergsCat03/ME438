```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the table
datasets = ["XRD", "Orbital", "Sine", "DFT"] * 3
models = ["SVR"] * 4 + ["GMM"] * 4 + ["NN"] * 4
train_rmse = [0.095, 0.126, 0.124, 0.119, 0.476, 0.270, 0.543, 0.392, 0.731, 1.013, 1.231, 1.482]
test_rmse = [0.087, 0.125, 0.179, 0.131, 1.323, 1.364, 1.150, 1.421, 1.874, 2.057, 2.012, 2.391]

# Unique values
unique_models = ["SVR", "GMM", "NN"]
unique_datasets = ["XRD", "Orbital", "Sine", "DFT"]

# Create figure and axis
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 12))
plt.style.use('seaborn')

# Colors for different datasets
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728']

# Plot Training RMSE
for i, dataset in enumerate(unique_datasets):
    dataset_train_rmse = []
    for j in range(3):  # For each model
        idx = unique_datasets.index(dataset) + j * 4
        dataset_train_rmse.append(train_rmse[idx])
    ax1.plot(unique_models, dataset_train_rmse, 'o-', label=dataset, color=colors[i], linewidth=2, markersize=8)

ax1.set_title('Training RMSE by Model and Dataset', fontsize=12, pad=15)
ax1.set_ylabel('Training RMSE')
ax1.grid(True, linestyle='--', alpha=0.7)
ax1.legend(title='Datasets', bbox_to_anchor=(1.05, 1), loc='upper left')

# Plot Testing RMSE
for i, dataset in enumerate(unique_datasets):
    dataset_test_rmse = []
    for j in range(3):  # For each model
        idx = unique_datasets.index(dataset) + j * 4
        dataset_test_rmse.append(test_rmse[idx])
    ax2.plot(unique_models, dataset_test_rmse, 'o-', label=dataset, color=colors[i], linewidth=2, markersize=8)

ax2.set_title('Testing RMSE by Model and Dataset', fontsize=12, pad=15)
ax2.set_xlabel('Models')
ax2.set_ylabel('Testing RMSE')
ax2.grid(True, linestyle='--', alpha=0.7)
ax2.legend(title='Datasets', bbox_to_anchor=(1.05, 1), loc='upper left')

# Adjust layout
plt.tight_layout()
plt.show()
```
